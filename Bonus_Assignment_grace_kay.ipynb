```
from newspaper import Article

details_np = pd.DataFrame(columns=["url", "authors", "publish_date", "text"])

for link in urls:
  details = []
  
  #Adding the url
  details.append(link)
  
  #Giving article link to newspaper
  michaeljackson = Article(link)
  
  #Dowloading the article
  michaeljackson.download()
  
  #Parsing the article
  michaeljackson.parse()
  
  #Retrieving name of the authors
  details.append(michaeljackson.authors)
    
  #Adding the publish date
  details.append(str(michaeljackson.publish_date))
  
  #Adding the text
  details.append(michaeljackson.text)
  
  details_np.loc[len(details_np)] = details
  
details_np
  
 
```
